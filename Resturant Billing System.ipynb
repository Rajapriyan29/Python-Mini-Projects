```python
menu ={
    'Pizza':150,
    'Burger':200,
    '7up':130,
    'Snacks':400
}
```

```python
bill=0
order=[]
```

```python
for item,price in menu.items():
    print(f"{item}-{price}")
```

```python
while True:
    item=input("Enter a item value")

    if item not in menu:
        print ("Not available in menu")
        continue

        qyt=int(input("Enter a qyt"))
        cost = menu[item] * qty
        bill += cost

        order_list.append([item, qty, cost])

        more = input("Add more items? (yes/no): ").lower()
        if more != "yes":
            break

gst = bill * 0.05
final_bill = bill + gst

print("\n========== RECEIPT ==========")
print("Item\t\tQty\tCost")
for order in order_list:
    print(f"{order[0]}\t\t{order[1]}\t₹{order[2]}")

print("-----------------------------")
print(f"Subtotal : ₹{bill:.2f}")
print(f"GST (5%) : ₹{gst:.2f}")
print(f"Total Bill: ₹{final_bill:.2f}")
print("=============================")
print("Thank You! Visit Again.")
```
