```python
import os
import shutil
```

```python
path='C:/CODE/Py File Organizers/'
```

```python
file_name=os.listdir(path)
```

```python
Folder_name=['csv file','image file']
```

```python
for loop in range(0,2):
    if not os.path.exists(path+Folder_name[loop]):
        print(path+Folder_name[loop])
        os.makedirs(path+Folder_name[loop])
```

```python
for files in file_name:
    if '.csv' and not os.path.exists(path+'csv file/' + files):
        shutil.move(path+files,path+'csv file/'+files)
  
```
