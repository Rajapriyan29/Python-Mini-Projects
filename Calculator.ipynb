```python
a=int(input("enter a 1st number"))
b=int(input("Enter a 2nd number"))
print("1.Add")
print("2.Subrate")
print("3.Multiple")
print("4.Division")
choice=input("choose 1-4")
if choice == 1:
    print("Result",a+b)
elif choice == 2:
    print("Result",a-b)
elif choice == 3:
    print("Result",a*b)
elif choice == 4:
    print("Result",a//b)
else:
    print("Invalid number")
    
```
