```python
import random
import string
```

```python
Word=int(input("How many letters in your password"))
```

```python
num=int(input("How many numbers in your password"))
```

```python
sym=int(input("How many symbols in your password"))
```

```python
letters=string.ascii_letters
```

```python
letters
```

```python
numbers=string.digits
```

```python
numbers
```

```python
symbols=string.punctuation
```

```python
symbols
```

```python
password=[]
```

```python
for i in range(1,Word+1):
    char=random.choice(letters)
    password+=char
    print(password)
```

```python
for i in range(1,num+1):
    dig=random.choice(numbers)
    password+=dig
    print(password)
```

```python
for i in range(1,sym+1):
    symb=random.choice(symbols)
    password.append(symb)
    print(password)
```

```python
random.shuffle(password)
```

```python
passw=''.join(password)
passw
```
