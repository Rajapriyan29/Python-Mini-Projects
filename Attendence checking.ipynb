```python
attendence={}

def add_student():
    name=input("Enter your name").lower()
    attendence[name]=[]
    print("Successfully added")
def mark():
    name = input("Enter a name")

    if name in attendence:
        status=input("present or Absent")
        attendence[name].append(status.upper())
        print("Attendence Marked")
    else:
        print("Attendence is not marked")
def view_attendence():
    for student, records in attendence.items():
        print(f"{student}: {records}")

while True:
    print("1.Add student\n")
    print("2.Mark attendence\n")
    print("3.View attendence\n")
    print("4.Exit")

    choice=input("enter your choice")

    if choice == "1":
        add_student()
    elif choice == "2":
        mark()
    elif choice == '3':
        view_attendence()
    elif choice == '4':
        break
    else:
        print("Invalid")
```
